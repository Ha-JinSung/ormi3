# 정규표현식

## 일반 문자열

```python
# 대소문자를 가려낼수 있다.
# ^hello는 처음에 나오는 hello
# hello$는 끝에 나오는 hello(hello^가 아님 주의!)
# h.llo는 h 다음에 어떤 문자든지 간에 1개의 문자 (영문, 특수문자, 숫자, 한글 등)가 오며 llow가 이어 나와야 합니다.
#
```

## 대괄호 (택 1)

```python
# [] 안에는 선택문자가 들어간다.
# 에시: h[eao]llo # [eao] 중 하나를 선택하기에 hello, hallow, hollo에서 택1 이다.
# h[eao]llo는 hello, hallo, hollo를 선택합니다. 3개 문자중 택1입니다.
# h[e!ao]llo는 hello, hallo, hollo, h!llo를 선택합니다. 3개 문자중 택1입니다.
# [a-z]
# [a-zA-Z]
# [0-9]
# [0-9]
# [a-zA-Z0-9]
# [가-힣]
# h1el222lo3wo123rl2d # [0-9]이면 0 ~ 9까지 숫자를 다 선택한다.
```

### 부정


```python
# ^[a-z] : 부정이 아니다 알파벳으로 시작하는 문자열
# [^a-z] : 이래야 부정이 적용된다 (a ~ z까지 문자 부정으로 처리하여 표기 안함) (대괄호 안에 넣어서 부정 쓰기)
# [a^o] : 이처럼 중간에 부정을 쓰면 사라진다. (^를 문자로 생각해 ^도 찾아낸다.) # 대괄호 안에서 첫문자열에 써야 적용
# h[^ao]llo : a, o 모두 하나씩 아닌것
# h[a^o]llo : a, ^, o 모두 하나씩 잡습니다. ^는 대괄호 열기 바로 다음에 나오게 해주세요.
```

## 수량자

```python
# _* : 앞에 있는 문자가 0개 ~ N개 # 어디서부터 시작하는지 중요함
# _+ : 앞에 있는 문자가 1개 ~ N개
# _? : 앞에 있는 문자가 0개 ~ 1개

#-------^ 위의 3개는 잘사용된다 중요

# {3} : 3개
# {3,} : 3개 이상
# {1,3} : 1개 ~ 3개

# _* : 앞에 있는 문자가 0개 ~ N개 ({0,})
# _+ : 앞에 있는 문자가 1개 ~ N개({1,})
# _? : 앞에 있는 문자가 0개 ~ 1개 ({0,1})
```

```python
# 이런 정규식은 인터넷 검색으로 나오니까 직접 만들지 말고 검색해서 쓰자. (전화번호, email등등)

# 왜 직접 만들지 않고 만들어진걸 쓰는가?

# 직접짜다 실수가 생기면 서비스에 cs 발생 및 크리티컬한 문제등등 발생한다.
# 다만 이 코드를 읽을줄은 알아야 한다.




# [0-9][0-9][0-9]-[0-9][0-9][0-9][0-9]-[0-9][0-9][0-9][0-9]
# 0[0-9]+[-. ]?[0-9]{3,4}[-. ]?[0-9]{3,4}
#
```

```python
# group1, group2
# group1 => key
# group2 => value
# ([a-zA-Z0-9@\.\-]+), ?([a-zA-Z0-9@\.\-]+)
# '[(name, leehojun), (age, 10), (height, 180), (email, paul-lab@naver.com)]'
```

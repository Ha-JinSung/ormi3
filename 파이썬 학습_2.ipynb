```python
# global을 하면 함수가 얼만큼 중첩되어 있던간에 최상단에 있는 a만 바라본다.
a = 10
def f():
    a = 100
    print(a)
    def ff():
        a = 1000
        print(a)
        def fff():
            a = 10000
            print(a)
        fff()
    ff()
f()
print(a)
```

```python
# global을 사용하지 않고 중첩된 함수 내에서 a를 출력하면 바로 위의 함수의 a로 출력
a = 10
def f():
    a = 100
    print(a)
    def ff():
        a = 1000
        print(a)
        def fff():
            print(a)
        fff()
    ff()
f()
print(a)
```

```python
# nonlocal
a = 10
def f():
    a = 100
    print(f'f a: {a}')
    def ff():
        a = 1000
        print(f'ff a: {a}')
        def fff():
            nonlocal a
            a = 100
            print(f'fff a: {a}')
        fff()
        print(f'ff a: {a}')
    ff()
f()
print(f'global a: {a}')
```

```python
# 암기 해두자
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

factorial(5)
'''
# 함수 호출     return
# factorial(n)  n * factorial(n-1)
# factorial(5)  5 * factorial(4) ==  5 * 24
# factorial(4)  4 * factorial(3) ==  4 * 6
# factorial(3)  3 * factorial(2) ==  3 * 2
# factorial(2)  2 * factorial(1) ==  2 * 1
# factorial(1)  1
'''
# 거꾸로 연산 후 다시 아래로 연산
# 비효율적 연산
```

```python
def reversed_str(s):
    if len(s) <= 1:
        return s
    return reversed_str(s[1:]) + s[0]

reversed_str('hello')

'''
함수 호출                           return
reversed_str(s)                     reversed_str(s[1:]) + s[0]
reversed_str('hello')               reversed_str(s[1:]) + s[0] == 'olle' + 'h'
reversed_str(4) ==('ello')          reversed_str(s[1:]) + s[0] == 'oll' + 'e'
reversed_str(3) ==('llo')           reversed_str(s[1:]) + s[0] == 'ol' + 'l'
reversed_str(2) ==('lo')            reversed_str(s[1:]) + s[0] == 'o' + 'l'
reversed_str(1) ==('o')             'o' == 'o'
'''
```

```python
data = [
    [2, 'hello world', 30],
    [3, 'hworl', 33],
    [5, 'hello wo', 35],
    [4, 'hello w', 31],
    [6, 'hello w', 39]
]

# 1번쨰 문제: 첫 번째 요소의 값을 정렬
sorted(data)

# [2, 'hello world', 30],
#  [3, 'hworl', 33],
#  [4, 'hello w', 31],
#  [5, 'hello wo', 35],
#  [6, 'hello w', 39]]

# 2번째 문제: 두 번째 요소의 문자열 길이를 기준을 정렬
sorted(data, key=lambda x:len(x[1]))

# [3, 'hworl', 33],
#  [4, 'hello w', 31],
#  [6, 'hello w', 39],
#  [5, 'hello wo', 35],
#  [2, 'hello world', 30]]

# 3번째 문제: 세 번째 요소 기준으로 정렬
sorted(data, key=lambda x:x[2])

# [2, 'hello world', 30],
#  [4, 'hello w', 31],
#  [3, 'hworl', 33],
#  [5, 'hello wo', 35],
#  [6, 'hello w', 39]]

# 4번째 문제: 첫 번째 요소와 세 번째 요소의 차가 큰 순으로 출력
sorted(data, key=lambda x: x[0] - x[2])

# 다른 답 sorted(data, key=lambda x: x[2]-x[0],reverse=True)

# [6, 'hello w', 39],
#  [3, 'hworl', 33],
#  [5, 'hello wo', 35],
#  [2, 'hello world', 30],
#  [4, 'hello w', 31]]
```

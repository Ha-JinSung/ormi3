# 매치 match

```python
text = 'Hello World'
match text:
    case 'Hello':
        print('Hello')
    case 'World':
        print('World')
    case _:
        print('No Match') # 결과 'No Match'
```

```python
# 이 코드는 작동 하지 않는다. JavaScript와 같은 곳에서는 아래 같은 형식을 사용하곤 이용한다.
text = 'Hello'
match text:
    case 'Hello':
    case 'Hello'and 'World':
        print('World')
    case 'World':
        print('World')
    case _:
        print('No Match')
```

```python
# 작동하는 코드입니다.
text = 'Hello'
match text:
    case 'Hello' | 'World':
        print('World')
    case _:
        print('No Match')
```

```python
def test(i):
    match i:
        case 1:
            print('One')
        case 2:
            print('Two')
        case 3 | 4:
            print('Three or Four')
        case _ if i > 10:
            print('Greater than ten')
        case _:
            print('Other')

test(1) # 출력: One
test(2) # 출력: Two
test(3) # 출력: Three or Four
test(4) # 출력: Three or Four
test(5) # 출력: Other
test(15) # 출력: Greater than ten
```

```python
def test(i):
    match i:
        case 1:
            print('One')
        case 2:
            print('Two')
        case _:
            if i > 2:
                print('Greater than two')
            else:
                print('Less than or equal to two')
test(1) # 출력: One
test(2) # 출력: Two
test(3) # 출력: Greater than two
test(4) # 출력: Greater than two
```

```python
# 문제 match를 사용해서 아래 코드 완성

def attack(attack_type, damege):
    match attack_type:
        case '물리공격':
            print(f'{damege}의 데미지를 입혔습니다.')
        case '마법공격':
            print(f'{damege}의 데미지를 입혔습니다.')
        case '정신공격':
            print(f'{damege}ms 만큼 상대방이 혼란을 느낍니다.')

attack('정신공격', 100)

# 물리공격, # 마법공격, # 정신공격

# 각 공격에 따라 아래의 값이 출력되어야 합니다.
# 물리공격, 마법공격 : '100의 데미지를 입혔습니다.'
# 정신공격 : '100ms 만큼 상대방이 혼란을 느낍니다.'
```

```python
import os

while True:
    userinput = input('>')
    if userinput == 'pwd':
        # print('현재 위치 출력')
        # print('\\'.join(__file__.split('\\')[:-1]))
        print(os.getcwd())
    elif userinput == 'dir' or userinput == 'ls':
        # print('현재 폴더에 폴더와 파일명 출력')
        print(os.listdir(os.getcwd()))
    elif userinput == 'exit':
        print('안녕히가세요.')
        break

# os 권한을 사용해서 사용자 동의 없이 무한 생성, 삭제등을 할수도 있기에 os 코딩은 조심
```

```python
i = 2
while i < 10:
    j = 1 # 별표 5개! while문은 중첩해서 사용할 때 초기화가 필요합니다!
    while j < 10:
        if j == 3:
            break
        print(f'{i} X {j} = {i*j}')
        j += 1
    i += 1
```

```python
# https://peps.python.org/pep-3136/
# StopIteration 에러를 이용한 무한루프 빠져 나오는 방법


try:
    for i in range(2, 10):
        for j in range(1, 10):
            if i*j > 6:
                raise StopIteration
            print (f'{i} X {j} = {i*j}')
except StopIteration:
    print ("Done")
```

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point((self.x + other.x), (self.y + other.y))

    def distance(self, other):
        dx = self.x - other.x
        dy = self.y - other.y
        return (dx**2 + dy**2)**0.5

dot1 = Point(10, 20)
dot2 = Point(20, 30)

dot3 = dot1 + dot2
dot3.x, dot3.y
dot3.distance(dot1)
```

```python
# 문제 3 BankAccount 클래스를 생성하세요. 이 클래스는 owner(계좌주 이름), balance(잔액) 속성을 가져야 합니다.
# 이 클래스는 deposit(amount) (입금) 및 withdraw(amount) (출금) 메서드를 가져야 합니다.
# 단, 출금 시 잔액보다 큰 금액을 출금하려고 하면 "잔액 부족" 메시지를 출력해야 합니다.

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return ''
        else:
            self.balance += amount
            return self.balance

    def withdraw(self, amount):
        if amount <= 0:
            return ''
        elif amount > self.balance:
            return '잔액 부족'
        else:
            self.balance -= amount
            return self.balance

account = BankAccount('owner', 10000)

print(account.deposit(100))
print(account.withdraw(100))
```

```python
# 문제 3 풀이

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액 부족")
        else:
            self.balance -= amount
            return self.balance

account = BankAccount('owner', 10000)

print(account.deposit(100))
print(account.withdraw(-100))
```

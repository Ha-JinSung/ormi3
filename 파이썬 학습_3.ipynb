```python
# 문제 1 : 다음 국영수 dict에 점수에 평균을 구하는 코드
국영수 = {'국어': 97, '영어': 38, '수학': 90}
print(sum(국영수.values()) / len(국영수))
```

# 얕은 복사와 깊은 복사

```python
import copy

a = [1, 2]
b = [3, 4]

aa = [10, 20, a]
bb = [30, 40, b]

aaa = [100, 200, aa]
bbb = [300, 400, bb]

l = [aaa, bbb]
ll = copy.deepcopy(l) # 깊은 복사

a[0] = 99

print(l)
print(ll)
```

```python

l = [[1, 2, 3], [4, 5, 6]]
ll = [i for i in l]
ll[0][0] = 1000

print(l)
print(ll)
```

```python
# 리스트 컴프리헨션으로 깊은 복사 구현 (완전하지 않는다.)
l = [[1, 2, 3], [4, 5, 6]]
ll = [i[:] for i in l]
ll[0][0] = 1000

print(l)
print(ll)
```

```python
# 위와 같은 방식으로 했을때 완전하지 않는 코드
a = [9, 99]

l = [[1, 2, a], [4, 5, 6]]
ll = [i[:] for i in l]
a[0] = 1000

print(l)
print(ll)
```

# 집합 (Set)

```python
# 1. 중복이 되지 않습니다.
# 2. 순서가 없습니다.
# 3. 교집합, 차집합, 합집합 등이 가능합니다.
```

```python
set('hello')
set(['licat', 'licat','licat', 'mura', 'binky']) # 중복된 사람 제거한다
s = {'licat', 'licat','licat', 'mura', 'binky'}
len(s)
```

```python
# 주의
t = {} # dict
t = () # tuple
t = (0) # int
t = (0,) # tuple
t = {0} # set
t = set()
type(t)
# t = tuple()
# type(t)
```

```python
# 합집합
s = {1, 1, 2, 3, 4, 4}
ss = {4, 5, 6}
# 중복된건 제거하고 합집합 한다
s | ss
```

```python
# 교집합

s = {1, 1, 2, 3, 4, 4}
ss = {4, 5, 6}
# 중복된게 4이다 교집합
s & ss
```

```python
# 차집합

s = {1, 1, 2, 3, 4, 4}
ss = {4, 5, 6}
# 중복된게 4이다 교집합
s - ss
```

```python
s = {1, 1, 2, 3, 4, 4}
ss = {4, 5, 6}

s + ss # error 이거 합집합 안된다 | 로 해야 합집합
```
